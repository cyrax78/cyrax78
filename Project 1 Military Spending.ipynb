```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import matplotlib
import matplotlib.ticker as mtick
%matplotlib inline

#construct dataframe, convert the unit of GDP and population to millions
militarySpend=pd.read_excel('Project 1 Data.xls',sheetname='Military Spending')
GDP=pd.read_excel('Project 1 Data.xls',sheetname='GDP')/1000000
population=pd.read_excel('Project 1 Data.xls',sheetname='Population')/1000000

#sort military spending by 2016 value (row value) and get top 15 countries in terms of total military spendings
militarySpend=militarySpend.sort_values(by=2016,ascending=False,axis=1)
militarySpendTop15=militarySpend.iloc[:,:15]
top15List=list(militarySpendTop15)

#only include top 15 countries in GDP and population data
GDPTop15=GDP[top15List]
populationTop15=population[top15List]

militarySpendTop15perCapita=militarySpendTop15.divide(populationTop15)
GDPTop15perCapita=GDPTop15.divide(populationTop15)
```

```python
##compare the data to the countries GDP
#total military spending to total GDP ratio in percentage
militarySpendTop15toGDP=militarySpendTop15.divide(GDPTop15)*100
militarySpendTop15toGDP=militarySpendTop15toGDP.sort_values(by=2016, ascending=False, axis=1)
militarySpendTop5toGDP=militarySpendTop15toGDP.iloc[:,:5].dropna(0) #remove the years that Russian Federation has no data

ax=militarySpendTop5toGDP.plot(figsize=(20,10),fontsize=14,legend=False)
plt.legend(loc=(0.45,0.7),fontsize=14)
ax.set_xlabel('Year',fontsize=14)
ax.set_ylabel('Military Spending to GDP Ratio in Percentage',fontsize=14)
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.2f}%'))
ax.set_title('Military Spending to GDP Ratio of Top 5 countries in G20',fontsize=20,x=0.5,y=0.95)
plt.savefig('Military Spending to GDP Ratio in Percentage.png')
```

```python
militarySpendTop5=militarySpendTop15.iloc[:,:5].dropna(0)
ax=militarySpendTop5.plot(figsize=(20,10),fontsize=14,legend=False)
plt.legend(loc=2,fontsize=14)
ax.set_xlabel('Year',fontsize=14)
#ax.set_ylabel('Military Spending in $MM',fontsize=14)
#ax.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda y,p: format(int(y),',')))
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
ax.set_title('Military Spending of Top 5 countries in G20 in $MM',fontsize=20,x=0.5,y=0.95)
```
